```javascript
var year_begin = 2018;
var year_end = year_begin + 3; // for lambda
var year_end2 = year_end;
// var year_end2 = year_end === 2019 ? year_end + 1 : year_end; // real filter

var imgcol = ee.ImageCollection('MODIS/061/MOD15A2H');
imgcol = imgcol.select('Lai_500m', 'FparExtra_QC')
  .filter(ee.Filter.calendarRange(year_begin, year_end2, "year"))
  .map(pkg_whit.qc_FparLai);
var ws = imgcol.select('w');
var lambda = pkg_whit.lambda_init(
  imgcol.select(0).filter(ee.Filter.calendarRange(year_begin, year_end, "year")))
  .divide(10);

var n = imgcol.size();
var dates = imgcol.aggregate_array('system:time_start');
// print(bands);
// Map.addLayer(lambda, { min: 0, max: 1000 }, 'lambda'); 
// print(lambda);
var options = { iters: 3, verbose: false, n: n };
var bands_sm = ee.List.sequence(1, options.iters);
// print(lambda, bands_sm);
// print("INPUTS:", imgcol, ws, lambda, options);
var imgcol_sm = pkg_whit.smooth_whit2(imgcol, ws, lambda, options);

// print('imgcol_sm', imgcol_sm);
// imgcol_sm
// Map.addLayer(imgcol_sm.limit(46), {}, 'smoothed');
```
