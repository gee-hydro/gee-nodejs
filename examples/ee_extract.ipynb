## Case1: Extract data async

```javascript
var options = pkg.guess_reduce_option(col);
// print("options", options);

var col = ee.ImageCollection("CAS/IGSNRR/PML/V2_v017");
var fc = ee.FeatureCollection("users/kongdd/shp/flux-212").select(['site', 'IGBP']).limit(10);

outdir = "data"
async function main() {  
  for (var year = 2000; year <= 2005; year++) {
    var _col = col.filter(ee.Filter.calendarRange(year, year, "year"));
    var f = outdir + "/data_" + year + ".json"
    console.log(f)
    
    // 等`ee_extract_col`执行完，再保存`write_json`
    // 实现异步任务
    var r = pkg.ee_extract_col(_col, fc, options);
    var tbl = pkg.sf_to_dict(r, true); //.getInfo()
    await pkg.ee_write_json(tbl, f)
    // write_json(tbl.getInfo(), f)  
  }
}

main()
```
